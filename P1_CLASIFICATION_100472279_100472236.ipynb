Antes de empezar a convertir nuestro problema de regresión en uno de clasificación tenemos que preparar nuestros datos.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import time
```

En primer lugar, y de acuerdo con el EDA realizado en el primer *notebook*, eliminaremos las variables que no se correspondan con la localización de Sotavento y las variables que decidimos que no aportaban nada relevante a nuestro dataset. Lo vemos a continuación:

```python
nombre_archivo = "wind_ava.csv.gz"
dataset = pd.read_csv(nombre_archivo, sep = ',')

columnas_drop= [col for col in dataset.columns if not (col.endswith('.13') or col == "datetime" or col == "energy")]
dataset = dataset.drop(columns=columnas_drop)

del dataset['lai_lv.13']
del dataset['u10.13']
del dataset['v10.13']

dataset['Fecha'] = pd.to_datetime(dataset['datetime'])
dataset['Año'] = dataset['Fecha'].dt.year
dataset['Mes'] = dataset['Fecha'].dt.month
dataset.set_index('Fecha', inplace=True)

del dataset['datetime']

dataset.drop(columns=['Año'])
```

A continuación, vamos a realizar la división de nuestros datos en *train* y *test*, respetando para ello el orden como hicimos en el problema de regresión.

```python
train_data = dataset.loc['2005-01-01':'2007-12-31']  # Datos de 2005, 2006 y 2007
test_data = dataset.loc['2008-01-01':'2009-12-31']   # Datos de 2008 y 2009

third_quartile= train_data["energy"].quantile(q=0.75)

print(train_data.describe())

# Definir una función para asignar los valores "alto" y "bajo"
def asignar_valor(row):
    if row["energy"] > third_quartile:
        return 'alto'
    else:
        return 'bajo'

# Aplicar la función a cada fila del DataFrame
#train_data.loc[:, 'Energy'] = train_data.apply(asignar_valor, axis=1)
train_data['Energy'] = train_data.apply(asignar_valor, axis = 1)
train_data = train_data.drop(columns=["energy"])

# Aplicar la función a cada fila del DataFrame
#test_data.loc[:, 'Energy'] = test_data.apply(asignar_valor, axis=1)
test_data['Energy'] = test_data.apply(asignar_valor, axis = 1)
test_data = test_data.drop(columns=["energy"])

X_train = train_data.drop(columns=["Energy"])
y_train = train_data["Energy"]
X_test = test_data.drop(columns=["Energy"])
y_test = test_data["Energy"]

print("Número de columnas en y_train:", y_train.shape)
print("Número de columnas en y_test:", y_test.shape)
print("Número de columnas en X_train:", X_train.shape)
print("Número de columnas en X_test:", X_test.shape)

print("Número de columnas en y_train:", y_train.shape)
print("Número de columnas en y_test:", y_test.shape)
```

## 1. Metodo de escalado ##

Igual que hicimos con el probrema de regresión, en primer lugar vamos a buscar qué metodo de escalado es mejor para nuestro problema de clasificación. Para ello

Para realizar esta búsqueda, configuraremos un *pipeline* para probar con *grid_search* los diferentes métodos de escalado *StandardScaler*, *RobustScaler* y *MinMaxScaler*.

```python
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score

# Definir el pipeline con los tres escaladores y el clasificador KNN
pipe = Pipeline([
    ('scaler', StandardScaler()),  # Primer escalador a evaluar
    ('knn', KNeighborsClassifier())  # Clasificador KNN
])

param_grid = {
    'scaler': [MinMaxScaler(), StandardScaler(), RobustScaler()],  # Escaladores a evaluar
}

# Inicializar TimeSeriesSplit para la validación cruzada
tscv = TimeSeriesSplit(n_splits=3)

# Inicializar GridSearchCV con el pipeline y la cuadrícula de parámetros
grid_search = GridSearchCV(pipe, param_grid, cv=tscv, scoring = 'balanced_accuracy')

# Entrenar GridSearchCV
grid_search.fit(X_train, y_train)

# Mostrar los mejores hiperparámetros encontrados
print("Mejor método de:")
print(grid_search.best_params_)

# Mostrar la precisión del modelo con los mejores hiperparámetros en el conjunto de prueba
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
precision = precision_score(y_test, y_pred, pos_label ='alto')
recall = recall_score(y_test, y_pred, pos_label ='alto')
f1 = f1_score(y_test, y_pred, pos_label ='alto')

print(f"Precisión del modelo en el conjunto de prueba: {precision}")
print(f"Recall del modelo en el conjunto de prueba: {recall}")
print(f"F1-score del modelo en el conjunto de prueba: {f1}")
```

Después de completar la búsqueda del mejor método de escalado, se determina que el método de escalado seleccionado es *StandardScaler*. Dicho scaler se utilizará de ahora en adelante para evaluar y entrenar los modelos.

```python
# Definir el scaler y aplicarlo a los datos de entrenamiento y prueba
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### KNN

Continuaremos con KNN (K vecinos más cercanos), que es un algoritmo simple pero efectivo en aprendizaje automático. Se basa en encontrar los puntos de datos más cercanos a un nuevo punto para realizar una predicción de a qué clase pertenece dicho punto.

Dado que KNN no tiene ningún atributo que permita controlar el desbalanceo de las clases de nuestros datos,  utilizaremos *RandomUnderSampler* para eliminar datos de la clase negativa (la clase mayoritaria).

```python
from imblearn.under_sampling import RandomUnderSampler

# Submuestreo aleatorio para equilibrar las clases
undersampler = RandomUnderSampler(random_state=42)
X_train_resampled, y_train_resampled = undersampler.fit_resample(X_train_scaled, y_train)
```

### Árbol de clasificación

Los árboles de clasificación son algoritmos que dividen los datos en subgrupos basados en características, formando una estructura de árbol donde cada nodo representa una característica y cada rama una decisión de clasificación. Son fáciles de interpretar y útiles para clasificar datos en diferentes categorías.

#### Arbol de clasificación con hiperparámetros por omisión

Después de experimentar con los diferentes modelos, así como con el ajuste de sus hiperparámetros, podemos extraer varias conclusiones importantes.

En primer lugar, observamos que la máquina de vector soporte con optimización de hiperparámetros (HPO) resulta ser el método más efectivo para nuestro problema de clasificación. Este modelo demuestra un rendimiento superior en términos de precisión, recall, F1-score y Balanced Accuracy en comparación con otros algoritmos evaluados.

Sin embargo, aunque la SVC con HPO se destaca en términos de rendimiento, es importante tener en cuenta que no es necesariamente el método más rápido. El modelo más rápido resulta ser el dummy, lo cual es comprensible dado que es un modelo extremadamente simple que no requiere ningún cálculo complicado.

En cuanto al ajuste de hiperparámetros, encontramos que para el caso de SVM, la optimización de hiperparámetros no conduce a una mejora significativa en comparación con los valores por omisión. Esto sugiere que es posible que hayamos experimentado sobreajuste al intentar ajustar los hiperparámetros de estos modelos.

Además, notamos que el modelo que más tiempo tarda en entrenarse es el SVM con HPO. Aunque el ajuste de hiperparámetros puede ser beneficioso para mejorar el rendimiento del modelo, es importante tener en cuenta el equilibrio entre el tiempo de ejecución y la mejora de los resultados.

A pesar de que, en el caso de SVM con HPO no haya habido una mejora de rendimiento, en otros modelos merce la pena invertir tiempo en las búsqueda de hiperparámetros para mejorarlo.
